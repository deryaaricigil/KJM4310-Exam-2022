```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_exercise1(beta, function_name, filename):
    #PLOTTING DETAILS 1/2
    size = 14;
    fig, ax = plt.subplots(figsize = (8, 5))

    #PLOTS
    plt.plot(beta, eval(function_name)(beta,1), label="J = 1", color='black', linewidth=1.5)
    plt.plot(beta, eval(function_name)(beta,2), label ="J = 2", color='grey', linewidth=1.5)
    plt.plot(beta, eval(function_name)(beta,3), label ="J = 3", color='mediumspringgreen', linewidth=1.5)

    #PLOTTING DETAILS 2/2
    #plt.title("Average Energy", loc = 'left', fontsize=20, fontname='Serif')
    plt.xlabel("$\\beta$", fontsize=size, fontname='Serif'); plt.ylabel("$\\langle E \\rangle $", fontsize=size, fontname='Serif')

    for tick in ax.get_xticklabels():
        tick.set_fontname("Serif")
    for tick in ax.get_yticklabels():
        tick.set_fontname("Serif")

    ax.tick_params(labelsize=size)
    plt.setp(ax.legend().texts, family='Serif', fontsize=size)
    plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)

    plt.savefig(filename+".pdf")
    plt.show()
```

## Exercise 1

```python
#c)

def E(beta, J):
    a = -3*J*np.sinh(3*J*beta) - 3*J*np.sinh(J*beta)
    b = np.cosh(3*J*beta) + 3*np.cosh(J*beta)
    return (a/b)

plot_exercise1(np.linspace(-3,4,200), "E", "1c")
```

```python
#d) 

def M(beta, J):
    q = 2*np.cosh(3*J*beta) + 6*np.cosh(J*beta)
    a = 8*np.cosh(J*beta) + 4*np.exp(3*J*beta)
    return a/q

plot_exercise1(np.linspace(-1,4,200), "M", "1d")
```

```python
#b)

x = np.linspace(-10,10,200)
size = 14; fig, ax = plt.subplots(figsize = (8, 5))

#PLOTS
plt.plot(x, np.sinh(x), '--', label="sinh(x)", color='cornflowerblue', linewidth=1.5)
plt.plot(x, np.cosh(x), '--', label="cosh(x)", color='mediumorchid', linewidth=1.5)
plt.plot(x, np.tanh(x), '|-', label="tanh(x)", color='black', linewidth=1.5, alpha=1.0)

#PLOTTING DETAILS
ax.set(xlim=(-3.5, 3.5), ylim=(-2, 2))
plt.xlabel("x", fontsize=size, fontname='Serif'); plt.ylabel("", fontsize=size, fontname='Serif')
for tick in ax.get_xticklabels():
    tick.set_fontname("Serif")
for tick in ax.get_yticklabels():
    tick.set_fontname("Serif")
ax.tick_params(labelsize=size); plt.setp(ax.legend().texts, family='Serif', fontsize=size)
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)
plt.savefig("2b.pdf")
plt.show()
```

```python
#d)

def M_ferro(B, N, J):
    a = N*np.sinh(B); b = np.sqrt( np.sinh(B)*np.sinh(B) - (np.exp(-4*J)))
    return a/b

def M_para(B,N):
    return N*np.tanh(B)

B = np.linspace(-30,30,200)
size = 14; fig, ax = plt.subplots(figsize = (8, 5))

#PLOTS
plt.plot(B, M_ferro(B,1,1), label="Ferromagnet, J = 1", color='black', linewidth=1.5)
plt.plot(B, M_ferro(B,1,5), '--', label="Ferromagnet, J = 5", color='lightgrey', linewidth=1.5)
plt.plot(B, M_para(B,1), label="Paramagnet", color='royalblue', linewidth=1.5)

#PLOTTING DETAILS
ax.set(xlim=(-3, 3), ylim=(-2.5, 2.5))
plt.xlabel("B", fontsize=size, fontname='Serif'); plt.ylabel("M", fontsize=size, fontname='Serif')
for tick in ax.get_xticklabels():
    tick.set_fontname("Serif")
for tick in ax.get_yticklabels():
    tick.set_fontname("Serif")
ax.tick_params(labelsize=size); plt.setp(ax.legend().texts, family='Serif', fontsize=size)
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)
plt.savefig("2d.pdf")
plt.show()
```

```python
#f)

T = np.linspace(-30,30,2000)
size = 14; fig, ax = plt.subplots(figsize = (8, 5))

def Cv(T,J,k):
    a = (J**2)/(k*(T**2))
    b = (np.cosh((1/(k*T))*J))**2
    return a/b

plt.plot(T,Cv(T,1,1), color='firebrick', label='J = 1')
plt.plot(T,Cv(T,3,1), '--', color='lightgrey', label='J = 3')

#PLOTTING DETAILS
plt.xlim([-4, 4]); plt.ylim([-0.02, 0.46])
plt.xlabel("T", fontsize=size, fontname='Serif'); plt.ylabel("$C_v$", fontsize=size, fontname='Serif')
for tick in ax.get_xticklabels():
    tick.set_fontname("Serif")
for tick in ax.get_yticklabels():
    tick.set_fontname("Serif")
ax.tick_params(labelsize=size); plt.setp(ax.legend().texts, family='Serif', fontsize=size)
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)
plt.savefig("2f.pdf")
plt.show()
```
